```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

file_loc = 'A:\Health Index\Data\Combined_data'
```

```python
os.listdir(file_loc)

countries_2018, LAD_2018, regions_2018 = [pd.read_csv(os.path.join(file_loc, i)) for i in os.listdir(file_loc)]

df_list = [countries_2018, LAD_2018, regions_2018]
```

```python
all_df = pd.DataFrame()

for df in df_list:
    df = df.drop(df.columns[[0, 1]], axis=1)
    all_df = pd.concat([all_df, df], axis=1)
```

```python
def regressionTestAndVisualise(dataframe, corr_method='pearson'):
    '''A function to run a multiple regression on a dataframe and plot a heatmap of those correlations.
    
    Choice of correlation method can be chosen within the function call.
    '''
    
    if corr_method == 'pearson':
        corr = dataframe.corr(method='pearson')
    else:
        corr = dataframe.corr(method=corr_method)

    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
    fig.colorbar(cax)
    ticks = np.arange(0,len(dataframe.columns),1)
    ax.set_xticks(ticks)
    plt.xticks(rotation=90)
    ax.set_yticks(ticks)
    ax.set_xticklabels(dataframe.columns)
    ax.set_yticklabels(dataframe.columns)
    plt.show()
    
    return corr
```

```python
correlation_df = regressionTestAndVisualise(all_df)

correlation_df
```
